```python
import pandas as pd

df = pd.read_csv(r"Titanic.csv")  # поменять путь на актуальный для вас

df.sample(5), df.columns, df.shape
```

**Упражнение 1.** Найдите средний возраст пассажиров

```python
df["Age"].mean()
```

**Упражнение 2.** Выведите имя самого молодого пассажира.

```python
df.loc[df["Age"].idxmin(), "Name"]
```

**Упражнение 3.** Посчитайте доли пассажиров по классам обслуживания (Pclass)

```python
df["Pclass"].value_counts(normalize=True)
```

**Упражнение 4.** Найдите средние возраста пассажиров по классам обслуживания.

```python
df.groupby("Pclass")["Age"].mean()
```

**Упражнение 5.** Удалите столбец с наибольшим числом пропусков.

```python
worst_col = df.isnull().sum().idxmax()
print(worst_col)
df = df.drop(columns=worst_col)
```

**Упражнение 6.** Удалите столбцы с наибольшим числом уникальных значений.

```python
max_uniq = df.nunique().max()
worst_cols = df.columns[df.nunique() == max_uniq].tolist()
print(worst_cols)
df = df.drop(columns=worst_cols)
```

**Упражнение 7.** Найдите доли выживших среди мужчин, среди женщин и среди всех пассажиров.

```python
print(df[df["Gender"] == "male"]["Survived"].mean())
print(df[df["Gender"] == "female"]["Survived"].mean())
print(df["Survived"].mean())
```

**Упражнение 8.** Найдите доли выживших по классам обслуживания.

```python
df.groupby("Pclass")["Survived"].mean()
```

**Упражнение 9.** Найдите доли выживших среди возрастных категорий 18-, 18-60, 60+.

```python
bins = pd.cut(df["Age"], bins=[0, 18, 60, 200], labels=["18-", "18-60", "60+"])
df.groupby(bins, observed=False)["Survived"].mean()
```

**Упражнение 10.** Любым из рассмотренных методов реализуйте left join, right join, inner join, outer join датафреймов df1 и df2 (объединение по полю A).

```python
df1 = pd.DataFrame(
    {
        "A": ["1", "2", "4", "7", "10"],
        "B": ["ab", "c", "bc", "ab", "efd"],
        "C": [4, 4, 3.5, 15, 5],
    }
)

df2 = pd.DataFrame(
    {
        "A": ["1", "2", "4", "7", "10"],
        "F": ["n22", "c1", "b8", "a589", "b95"],
        "D": [100, 400, 30, 150, 50],
    }
)
```

```python
print("left join")
print(df1.merge(df2, on="A", how="left"))

print("\nright join")
print(df1.merge(df2, on="A", how="right"))

print("\ninner join")
print(df1.merge(df2, on="A", how="inner"))

print("\nouter join")
print(df1.merge(df2, on="A", how="outer"))
```
